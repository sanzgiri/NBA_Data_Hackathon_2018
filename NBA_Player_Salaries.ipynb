Using Spotrac data

```
def get_salaries_by_team_year_spotrac(team, year):
  
  url = 'https://www.spotrac.com/nba/{0}/cap/{1}/'.format(team, year)
  df = pd.read_html(url)
  return df[0]
```

```
get_salaries_by_team_year_spotrac('atlanta-hawks', 2017)
```

```
def get_salaries_by_season_hoopshype(season):
  
  url = 'https://hoopshype.com/salaries/players/{0}/'.format(season)
  df = pd.read_html(url)
  return df[0].drop(['Unnamed: 0'], axis=1)
```

```
get_salaries_by_season_hoopshype('2017-2018')
```
